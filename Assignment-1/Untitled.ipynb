```python
list_words = list()
with open('word_cloud/98-0.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        list_words.append(word for word in line.strip().split(' '))
```

```python
words_list = []
for line in lines:
    words = [word.strip(',.?!()').lower() for word in line.strip().split(' ') if not word == ''  ]
    words_list.extend(words)
```

```python
'the' in words_list
```

```python
len(words_list)
```

```python
word_dict = dict()

for word in words_list:
    if word in word_dict:
        word_dict[word] +=1
    else:
        if word not in stopwords_list:
            word_dict[word] = 1
            
        
```
